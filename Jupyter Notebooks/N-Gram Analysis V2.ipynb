```python
import nltk
import pandas as pd
from nltk import bigrams
from nltk import trigrams
nltk.download('stopwords') 

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('english')) # set of English stopwords

column = [0, 2]
#text = str(pd.read_excel(r'c:\Users\evaga\OneDrive\Documents\GitHub\SwEng-Group-1-NER\Datasets\Dell Output PROCESSED.xlsx', usecols= column ))
file = open(r'c:\Users\evaga\OneDrive\Documents\GitHub\SwEng-Group-1-NER\Datasets\Dell Output PROCESSED')
text = str(file.read)
#split the texts into tokens
print(text)
tokens = nltk.word_tokenize(text)
tokens = [token.lower() for token in tokens if len(token) > 1] #same as unigrams
filtered_tokens = [token for token in tokens if token not in stop_words] # remove the stopwords

bi_tokens = bigrams(filtered_tokens)
tri_tokens = trigrams(filtered_tokens)

bi_tokens = list(bigrams(filtered_tokens))
tri_tokens = list(trigrams(filtered_tokens))

# print trigrams count
#print ([(item, tri_tokens.count(item)) for item in sorted(set(tri_tokens))])
```

```python

from collections import Counter
ngram_counts = Counter(filtered_tokens)

import matplotlib.pyplot as plt

min_count = 2 # change this to the desired minimum frequency threshold
ngram_counts_filtered = {ng:freq for ng, freq in ngram_counts.items() if freq >= min_count}
ngram_labels = [ng for ng in ngram_counts_filtered.keys()]
ngram_freqs = [freq for freq in ngram_counts_filtered.values()]


plt.bar(ngram_labels, ngram_freqs)
plt.xticks(rotation=90)
plt.xlabel('N-gram')
plt.ylabel('Frequency')
plt.title('N-gram Frequencies (min count = {})'.format(min_count))
plt.show()
```
